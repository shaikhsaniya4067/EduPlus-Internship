```python
# 1. Create a dictionary with 5 students and their marks
students = {
    "Aman": 85,
    "Riya": 92,
    "Neha": 78,
    "Arjun": 88,
    "Karan": 67
}
print(students)
```

```python
# 2. Print the value of a specific student using key
print("Marks of Riya:", students["Riya"])
```

```python
# 3. Add a new student with marks
students["Sonal"] = 90
```

```python
# 4. Update the marks of an existing student
students["Neha"] = 82
```

```python
# 5. Delete a student from the dictionary
del students["Karan"]
```

```python
# 6. Print all student names using keys()
print("Student Names:", students.keys())
```

```python
# 7. Print all student marks using values()
print("Student Marks:", students.values())
```

```python
# 8. Print all student names with their marks using items()
print("Students and Marks:")
for name, marks in students.items():
    print(name, ":", marks)
```

```python
# 9. Check if a specific student is in the dictionary
print("Is Aman present?", "Aman" in students)
```

```python
# 10. Check if a specific mark is present in the dictionary values
print("Is 90 present?", 90 in students.values())
```

```python
# 11. Use get() to fetch marks safely
print("Marks of Rahul:", students.get("Rahul", "Student not found"))
```

```python
# 12. Create a dictionary of employees with departments
employees = {
    "Anita": "HR",
    "Rohit": "IT",
    "Meena": "Finance",
    "Suresh": "IT"
}
```

```python
# 13. Access department of a specific employee
print("Rohit's Department:", employees["Rohit"])
```

```python
# 14. Add a new employee
employees["Vikas"] = "Marketing"
```

```python
# 15. Change department of an existing employee
employees["Meena"] = "HR"
```

```python
# 16. Remove an employee
del employees["Suresh"]
```

```python
# 17. Sort by employee names (ascending)
print("Sorted by names (ASC):", dict(sorted(employees.items())))
```

```python
# 18. Sort by department (ascending)
print("Sorted by department (ASC):",
dict(sorted(employees.items(), key=lambda x: x[1])))
```

```python
# 19. Sort by employee names (descending)
print("Sorted by names (DESC):",
      dict(sorted(employees.items(), reverse=True)))
```

```python
# 20. Sort by department (descending)
print("Sorted by department (DESC):",
      dict(sorted(employees.items(), key=lambda x: x[1], reverse=True)))
```

```python
# 21. Find all keys that have a specific value
value_to_find = "HR"
keys_with_value = [k for k, v in employees.items() if v == value_to_find]
print("Employees in HR:", keys_with_value)
```

```python
# 22. Count students who scored above 80
count_above_80 = sum(1 for m in students.values() if m > 80)
print("Students scoring above 80:", count_above_80)
```

```python
# 23. Nested dictionary for 3 students
student_details = {
    "Aman": {"age": 20, "grade": "A", "department": "CS"},
    "Riya": {"age": 21, "grade": "A+", "department": "IT"},
    "Neha": {"age": 19, "grade": "B", "department": "ECE"}
}
```

```python
# 24. Access grade of a student
print("Riya's Grade:", student_details["Riya"]["grade"])
```

```python
# 25. Update department of a student
student_details["Neha"]["department"] = "CS"
```

```python
# 26. Loop and print student names with grades
print("Student Grades:")
for name, info in student_details.items():
    print(name, ":", info["grade"])
```

```python
# 27. Merge two dictionaries
students2 = {"Rahul": 75, "Pooja": 89}
merged_students = students | students2
print("Merged Dictionary:", merged_students)
```

```python
# 28. Clear all entries in a dictionary
students2.clear()
print("Cleared Dictionary:", students2)
```

```python
# 29. Copy a dictionary and modify the copy
students_copy = students.copy()
students_copy["Aman"] = 95
print("Original:", students)
print("Copy:", students_copy)
```
