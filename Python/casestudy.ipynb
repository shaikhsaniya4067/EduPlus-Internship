```python
# CaseStudyA
employees = { # 1.Display all employees who know both ‘Python’ and ‘SQL’.
    "Alice": ["Python", "SQL", "Java"],
    "Bob": ["C", "C++"],
    "Charlie": ["Python", "SQL"]
}
print("Employees who know both Python and SQL:")
for name, skills in employees.items():
    if "Python" in skills and "SQL" in skills:
        print(name)  
```

```python
unique_skills = set() # 2.Find and print the unique skills available across all employees (using sets).
for skills in employees.values():
    unique_skills.update(skills)
print("Unique skills across all employees:")
print(unique_skills)
```

```python
emp_name = input("Enter employee name: ") # 3.Identify how many employees share at least one skill with a specific employee entered by the user.
if emp_name not in employees:
    print("Employee not found.")
else:
    # Convert selected employee skills to set
    emp_skills = set(employees[emp_name])
    count = 0
    print("\nEmployees sharing at least one skill:")
    for name, skills in employees.items():
        if name != emp_name:
            # Check for common skills
            if emp_skills.intersection(skills):
                print(name)
                count += 1
    print(f"\nTotal employees sharing at least one skill with {emp_name}: {count}")
```

```python
### CaseStudyB
# Product Inventory Analyzer (Warehouse Comparison)
# Function to take product details for a warehouse
def input_warehouse_data(warehouse_name):
    n = int(input(f"\nEnter number of products in {warehouse_name}: "))
    products = []

    for i in range(n):
        print(f"\nEnter details for product {i+1}:")
        pid = input("Product ID: ")
        pname = input("Product Name: ")
        qty = int(input("Quantity: "))

        products.append({
            "id": pid,
            "name": pname,
            "quantity": qty
        })
    return products
# Input data for both warehouses
warehouse1 = input_warehouse_data("Warehouse 1")
warehouse2 = input_warehouse_data("Warehouse 2")
```

```python
#1. Common products (by name)
names_w1 = {p["name"] for p in warehouse1}
names_w2 = {p["name"] for p in warehouse2}

common_products = names_w1.intersection(names_w2)

print("\nCommon Products in Both Warehouses:")
if common_products:
    for name in common_products:
        print(name)
else:
    print("No common products found.")
```

```python
# 2. Unique products across both warehouses
all_unique_products = names_w1.union(names_w2)

print("\nAll Unique Products Across Both Warehouses:")
for name in all_unique_products:
    print(name)
```

```python
# 3. Combine and sort by quantity (descending)

combined_products = warehouse1 + warehouse2
sorted_products = sorted(combined_products, key=lambda x: x["quantity"], reverse=True)

print("\nCombined Product List (Sorted by Quantity - Descending):")
print("ID\tName\t\tQuantity")
for p in sorted_products:
    print(f"{p['id']}\t{p['name']}\t\t{p['quantity']}")
```

```python
# CaseStudyC
# University Course Registration System
# Input student details
n = int(input("Enter number of students: "))
students = []
for i in range(n):
    print(f"\nEnter details for student {i+1}:")
    name = input("Student Name: ")
    courses = tuple(input("Enter courses (comma-separated): ").split(","))
    # Remove extra spaces from course names
    courses = tuple(course.strip() for course in courses)
    students.append((name, courses))
```

```python
# 1. Students enrolled in both Data Science and AI
print("\nStudents enrolled in both Data Science and AI:")
found = False
for name, courses in students:
    if "Data Science" in courses and "AI" in courses:
        print(name)
        found = True
if not found:
    print("No student enrolled in both courses.")
```

```python
# 2. Total number of unique courses
all_courses = set()
for _, courses in students:
    all_courses.update(courses)
print("\nTotal number of unique courses offered:", len(all_courses))
```

```python
# 3. Courses that no student has repeated
# (courses taken by exactly one student)
course_count = {}

for _, courses in students:
    for course in courses:
        course_count[course] = course_count.get(course, 0) + 1

non_repeated_courses = [course for course, count in course_count.items() if count == 1]

print("\nCourses that no student has repeated:")
if non_repeated_courses:
    for course in non_repeated_courses:
        print(course)
else:
    print("No such courses found.")
```

```python
# CaseStudyD
# Customer Purchase Insights (E-commerce Scenario)
# Input customer purchase details
n = int(input("Enter number of customers: "))
purchases = []
for i in range(n):
    print(f"\nEnter details for customer {i+1}:")
    name = input("Customer Name: ")
    products = tuple(input("Enter products purchased (comma-separated): ").split(","))
    # Remove extra spaces
    products = tuple(product.strip() for product in products)
    purchases.append((name, products))
```

```python
# 1. Total number of unique products
all_products = set()
for _, products in purchases:
    all_products.update(products)

print("\nTotal number of unique products bought:",
      len(all_products))
```

```python
# 2. Customers who purchased both Laptop and Headphones
print("\nCustomers who purchased both Laptop and Headphones:")
found = False
for name, products in purchases:
    if "Laptop" in products and "Headphones" in products:
        print(name)
        found = True

if not found:
    print("No such customers found.")
```

```python
# 3. Most frequently purchased product(s)
product_count = {}
for _, products in purchases:
    for product in products:
        product_count[product] = product_count.get(product, 0) + 1
max_count = max(product_count.values())
most_frequent = [p for p, c in product_count.items() if c == max_count]
print("\nMost frequently purchased product(s):")
for product in most_frequent:
    print(f"{product} (purchased {max_count} times)")
```

```python
# CaseStudyE
# Travel Planner (Data Deduplication Challenge)
# Input booking details
n = int(input("Enter number of booking records: "))
bookings = []

for i in range(n):
    print(f"\nEnter details for booking {i+1}:")
    name = input("Traveler Name: ")
    destination = input("Destination: ")
    date = input("Travel Date (DD-MM-YYYY): ")

    # Store each record as a tuple
    bookings.append((name, destination, date))
```

```python
# 1. Remove duplicate records using set
unique_bookings = set(bookings)
print("\nUnique Booking Records:")
for record in unique_bookings:
    print(record)
```

```python
# 2. Display all unique destinations
unique_destinations = {destination for _, destination, _ in unique_bookings}
print("\nUnique Destinations:")
for dest in unique_destinations:
    print(dest)
```

```python
# 3. Top 3 most visited destinations
destination_count = {}
for _, destination, _ in unique_bookings:
    destination_count[destination] = destination_count.get(destination, 0) + 1
# Sort destinations by frequency (descending)
sorted_destinations = sorted(destination_count.items(),
                             key=lambda x: x[1],
                             reverse=True)
print("\nTop 3 Most Visited Destinations:")
for dest, count in sorted_destinations[:3]:
    print(f"{dest} - {count} bookings")
```

```python
# CaseStudyF
# Student Attendance Analyzer
attendance = []
# Input attendance for 5 days
for day in range(1, 6):
    print(f"\nEnter students present on Day {day}:")
    students = set(input("Enter names (comma-separated): ").split(","))
    # Remove extra spaces
    students = {student.strip() for student in students}
    attendance.append(students)
```

```python
# 1. Students present every day
present_every_day = attendance[0]

for day_attendance in attendance[1:]:
    present_every_day = present_every_day.intersection(day_attendance)

print("\nStudents present every day:")
if present_every_day:
    for student in present_every_day:
        print(student)
else:
    print("No student was present every day.")
```

```python
# 2. Students absent on at least one day
all_students = set()

for day_attendance in attendance:
    all_students = all_students.union(day_attendance)

absent_at_least_once = all_students - present_every_day

print("\nStudents absent on at least one day:")
for student in absent_at_least_once:
    print(student)
```

```python
# 3. Total unique students
print("\nTotal unique students across all five days:",
      len(all_students))
```

```python
# CaseStudy G: Movie Recommendation Tracker

# Input user details
n = int(input("Enter number of users: "))
users = []

for i in range(n):
    print(f"\nEnter details for user {i+1}:")
    name = input("User Name: ")
    movies = tuple(input("Enter favorite movies (comma-separated): ").split(","))

    # Remove extra spaces
    movies = tuple(movie.strip() for movie in movies)

    users.append((name, movies))
```

```python
# 1. Users sharing at least two movies
print("\nUsers who share at least two movies in common:")

found = False
for i in range(len(users)):
    for j in range(i + 1, len(users)):
        common_movies = set(users[i][1]).intersection(users[j][1])
        if len(common_movies) >= 2:
            print(f"{users[i][0]} and {users[j][0]} -> {common_movies}")
            found = True

if not found:
    print("No users share at least two movies.")
```

```python
# 2. All unique movies
all_movies = set()

for _, movies in users:
    all_movies.update(movies)

print("\nAll unique movies in the database:")
for movie in all_movies:
    print(movie)
```

```python
# 3. Movies liked by only one user
movie_count = {}

for _, movies in users:
    for movie in movies:
        movie_count[movie] = movie_count.get(movie, 0) + 1

unique_user_movies = [m for m, c in movie_count.items() if c == 1]

print("\nMovies marked as favorite by only one user:")
if unique_user_movies:
    for movie in unique_user_movies:
        print(movie)
else:
    print("No such movies found.")
```

```python
# CaseStudy H: Sports Event Participation Summary

# Input participant details
n = int(input("Enter number of participants: "))
participants = []

for i in range(n):
    print(f"\nEnter details for participant {i+1}:")
    name = input("Participant Name: ")
    sports = tuple(input("Enter sports played (comma-separated): ").split(","))

    # Remove extra spaces
    sports = tuple(sport.strip() for sport in sports)

    participants.append((name, sports))
```

```python
# 1. Participants playing both Cricket and Football
print("\nParticipants playing both Cricket and Football:")
found = False
for name, sports in participants:
    if "Cricket" in sports and "Football" in sports:
        print(name)
        found = True

if not found:
    print("No participant plays both Cricket and Football.")
```

```python
# 2. Total number of unique sports
all_sports = set()

for _, sports in participants:
    all_sports.update(sports)

print("\nTotal number of unique sports being played:",
      len(all_sports))
```

```python
# 3. Sports with highest participation count
sport_count = {}

for _, sports in participants:
    for sport in sports:
        sport_count[sport] = sport_count.get(sport, 0) + 1

max_count = max(sport_count.values())
most_popular_sports = tuple(
    sport for sport, count in sport_count.items()
    if count == max_count
)
print("\nSports with the highest participation count:")
print(most_popular_sports)
```
