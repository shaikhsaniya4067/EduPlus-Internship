```python
# Q1: Create a DataFrame with the following e-commerce sales data:
# Columns: OrderID, CustomerID, Product, Category, Quantity, Price, OrderDate, Rating, Review
# Create at least 50 sample records with realistic data including dates from 2023-2024
# Display the first 10 and last 5 records using head() and tail()

import pandas as pd
import numpy as np

# Create sample e-commerce data (50 records)
data = {
    "OrderID": range(1001, 1051),
    "CustomerID": range(501, 551),
    "Product": [
        "Laptop", "Mobile", "Headphones", "Keyboard", "Mouse",
        "Tablet", "Smartwatch", "Charger", "Camera", "Printer"
    ] * 5,
    "Category": [
        "Electronics", "Electronics", "Accessories", "Accessories", "Accessories",
        "Electronics", "Wearables", "Accessories", "Electronics", "Electronics"
    ] * 5,
    "Quantity": [
        1, 2, 1, 1, 2,
        1, 1, 3, 1, 1
    ] * 5,
    "Price": [
        75000, 30000, 2500, 1500, 900,
        40000, 12000, 800, 55000, 18000
    ] * 5,
    "OrderDate": pd.date_range(start="2023-01-01", periods=50, freq="W"),
    "Rating": [
        5, 4, 4, 3, 5,
        4, 5, 3, 4, 4
    ] * 5,
    "Review": [
        "Excellent", "Very Good", "Good", "Average", "Excellent",
        "Good", "Excellent", "Average", "Very Good", "Good"
    ] * 5
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Display the first 10 
print("First 10 records:")
print(df.head(10))
```

```python
# Display last 5 records
print("Last 5 records:")
print(df.tail(5))
```

```python
# Q2: Use info() and describe() to get a comprehensive overview of the dataset
# Identify data types, missing values, and statistical summary
# Print the column names using columns attribute and shape of the dataset

print("Data Info:")  # info()
print(df.info())

print("Summary of data:")  # Describe()
print(df.describe())

print("Data Types:",df.dtypes) # datatypes

print("Missing Values:", df.isnull()) # check missing values

print("\nColumn Names:", df.columns) # print column names

print("\nDataset Shape:", df.shape)  # print dataset shape
```

```python
# Q3: Introduce 10% missing values randomly in Rating and Review columns
# Detect missing values and display the count of missing values per column
# Fill missing ratings with the median rating and missing reviews with "No review provided"

df.loc[1,'Rating'] = None  # Introduce missing values in Rating columns
df.loc[2,'Rating'] = None
df.loc[3,'Rating'] = None
df.loc[4,'Rating'] = None
df.loc[5,'Rating'] = None
print("with missing data:\n",df['Rating'])

# Detect missing values and display the count of missing values per column
print("\nDetect Missing values in Rating")
print(df['Rating'].isnull().sum())

# Fill missing ratings with the median rating
df['Rating']=df['Rating'].fillna(df['Rating'].mean()) 
print("\nFill Missing Rating", df['Rating'])

df.loc[1,'Review'] = None  # Introduce missing values in Review coulmns
df.loc[2,'Review'] = None
df.loc[3,'Review'] = None
df.loc[4,'Review'] = None
df.loc[5,'Review'] = None
print("\nwith missing data:",df['Review'])

# Detect missing values and display the count of missing values per column
print("\nDetect Missing values in Review")
print(df['Review'].isnull().sum())

# Missing reviews with "No review provided"
df['Review'].fillna("No review provided", inplace=True)
print(df['Review'])
```

```python
# Q4: Create a new column 'TotalAmount' by multiplying Quantity and Price
# Add another column 'DiscountedPrice' that applies 10% discount on orders above $100
# Use conditional selection to filter orders where TotalAmount > $150

# Create TotalAmount column
df['TotalAmount'] = df['Quantity'] * df['Price']

# Apply 10% discount if TotalAmount > 30000
df['DiscountedPrice'] = df['TotalAmount'].where(
    df['TotalAmount'] <= 30000,
    df['TotalAmount'] * 0.9
)

# Filter orders where TotalAmount > 35000
filter_df = df[df['TotalAmount'] > 35000]

# Display results
print("Full DataFrame:\n")
print(df)

print("\nFiltered Orders (TotalAmount > 35000):\n")
print(filter_df)
```

```python
# Q5: Select only the columns: Product, Category, TotalAmount, and Rating
# Filter and display all orders from the 'Electronics' category with Rating >= 4

select_cols = df[["Product", "Category", "TotalAmount", "Rating"]] #Select only the columns: Product, Category, TotalAmount, and Rating
electronics_rating = select_cols[
    (select_cols["Category"] == "Electronics") &
    (select_cols["Rating"] >= 4)
]

print("Electronics Orders with Rating >= 4:\n")
print(electronics_rating)
```

```python
# Q6: Use boolean indexing to find all orders where:
# - Quantity > 2 AND Price < 50 OR Rating == 5
# Display the count of such orders

orders = df[
    ((df["Quantity"] > 2) & (df["Price"] < 50)) |
    (df["Rating"] == 5)
]
print("Count of such ordes:")
print(orders.shape)
```

```python
# Q7: Sort the entire DataFrame by OrderDate (descending) and then by TotalAmount (descending)
# Create a new column 'SalesRank' that ranks products by TotalAmount in descending order
df_sorted = df.sort_values(
    by=["OrderDate", "TotalAmount"],
    ascending=[False, False]
)
df_sorted["SalesRank"] = df_sorted["TotalAmount"].rank(
    method="dense",
    ascending=False
)
print("Sorted DataFrame with Sales Rank:\n")
print(df_sorted.head(10))
```

```python
# Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order
top_products= (
    df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)
)
print(top_products)
```
