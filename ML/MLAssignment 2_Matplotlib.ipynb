```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.
import matplotlib.pyplot as plt
import numpy as np
# Creating data for plotting 
x = np.linspace(0,10)
y = 2 * x + 3

# create a simple Line plot
plt.plot(x,y)

# Add title and Labels
plt.title("Simple Line Plot of y=2x+3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
# Q2: Plot sin(x) and cos(x) on the same graph with different colors, line styles, and labels, and display a legend.
x = np.linspace(0, 10)
y1= np.sin(x)
y2 = np.cos(x)

# Plot multiple lines with customization
plt.plot(x,y1, color='purple', linestyle='--', marker='o', label='sin(x)')
plt.plot(x,y2, color='green', linestyle='-', marker='x', label='cos(x)')

plt.title("Line Plot of Same Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Add gird and legend
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Q3: Create a bar chart showing the sales of four products: A, B, C, and D with random values.
products=['A','B','C','D']
values=[20,25,30,40]
plt.bar(products, values, color='cyan')
plt.title("Bar Chart")
plt.xlabel("products")
plt.ylabel("values")
plt.show()
```

```python
# Q4: Generate a scatter plot with random x and y values between 0 and 1, and use different colors for points.
# Create random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# Create scatter plot
plt.scatter(x, y, c=colors, cmap='viridis')

# Add title and labels
plt.title("Scatter Plot with Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add colorbar
plt.colorbar()

# Display plot
plt.show()
```

```python
# Q5: Plot a histogram for 1000 random numbers generated from a normal distribution.
# Random data for histogram
data = np.random.randn(1000) #1000 random numbers

# Plot histogram
plt.hist(data, bins=30, color='magenta', edgecolor='black', alpha=0.7)

plt.title("Histogram")
plt.xlabel("Data Range")
plt.ylabel("Fregquency")

plt.show()
```

```python
# Q6: Draw a pie chart representing the percentage of five fruits: Apple, Banana, Mango, Grapes, and Orange.
# Data for a pie chart
labels = ['Apple','Banana','Mango','Grapes','Orange']
sizes = [30,20,25,25,35]
colors= ['red','yellow','pink','orange','green']

# craete pie chart
plt.pie(sizes, labels=labels,colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)

plt.title("Pie Chart")
plt.show()
```

```python
# Q7: Create two subplots in one figure  one showing a sine wave and the other showing a cosine wave.
# Create data
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

# Craete subplots
plt.figure(figsize=(10,5))

# First subplot
plt.subplot(1,2,1)  # (rows, column,index)
plt.plot(x, y1, 'r--')
plt.title("sin(x)")

# Second subplot
plt.subplot(1,2,2)  
plt.plot(x, y2, 'b-')
plt.title("cos(x)")

plt.suptitle("Two Subplots")  # overall title
plt.tight_layout()
plt.show()
```

```python
# Q8: Customize a line plot by adding title, axis labels, grid lines, and saving it as 'line_plot.png'.
x = np.linspace(0, 5)
y = np.linspace(1,10)

plt.plot(x, y)
plt.title("Customize Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save plot as PNG file
plt.savefig("Matplotlib-Library-customize_plot.png", dpi=300, bbox_inches='tight')
plt.grid(True)
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.
#create figure and subplot
fig, axs= plt.subplots(2,2, figsize=(10,8))

#Line Plot
x = np.linspace(0, 10, 100)
axs[0,0].plot(x, np.sin(x), 'r')
axs[0,0].set_title("Sine Wave")

# Bar Chart
categories = ['A','B','C', 'D']
values = [10,15, 7, 12]
axs[0,1].bar(categories, values, color='teal')
axs[0,1].set_title("Bar Chart")

# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
axs[1,1].scatter(x, y, color='orange')
axs[1,1].set_title("Scatter Plot")

# Histogram
data = np.random.randn(1000)
axs[1,0].hist(data, bins=25, color='purple', edgecolor='black')
axs[1,0].set_title("Histogram")

plt.suptitle("Matplotlib Dashboard Example", fontsize=14)
plt.tight_layout()

# Save plot as Png File
plt.savefig("Matplotlib.Dashboard.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Q10: Create a line plot for y = e^x for x between 0 and 5, and change the color, marker, and line style.
x = np.linspace(0, 5)
y = np.exp(x)

plt.plot(x, y, color='green', marker='o', linestyle='--')
plt.title("Exponential Function")
plt.xlabel("X")
plt.ylabel("exp(X)")

plt.show()
```
